```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
src_img = cv2.imread("./week1_homework.png")
src_img1 = cv2.cvtColor(src_img,cv2.COLOR_BGR2RGB)
plt.imshow(src_img1)
```

```python
#高斯滤波效果
dst_img_aver = cv2.GaussianBlur(src_img1,(11,11),1,0)
plt.imshow(dst_img_aver)
```

```python
#特定滤波设计
#老照片---- 彩色变黑白
src_img_gray = cv2.cvtColor(src_img1,cv2.COLOR_RGB2GRAY)
#plt.imshow(src_img_gray,cmap='gray')

ret, binary = cv2.threshold(src_img_gray, 127, 255, cv2.THRESH_BINARY)
print(ret)
#cv2.imshow("binary", binary)
plt.imshow(binary,cmap='gray')
```

```python
#均值滤波
kernel = np.ones((3,3),np.float32)/9
# kernel=-kernel
# kernel[0,:]=[-1,-1,-1]
# kernel[1,:]=[-1,8,-1]
# kernel[2,:]=[-1,-1,-1]
plt.imshow(src_img_gray,cmap='gray')
result = cv2.filter2D(src_img_gray,-1,kernel)
print(result[100,100])
print(kernel)
plt.imshow(result,cmap='gray')
```
